```python
import pandas as pd

csv_file_path = "C:/Users/suhai/Downloads/Employee_HR.csv"
df = pd.read_csv(csv_file_path)
print(df.head())
```

```python
hr_marketing_salaries = df[(df['Department'] == 'HR') | (df['Department'] == 'marketing')]

salary_variance = hr_marketing_salaries['Salary_INR'].var()
salary_std_deviation = hr_marketing_salaries['Salary_INR'].std()

q1 = hr_marketing_salaries['Salary_INR'].quantile(0.25)
q3 = hr_marketing_salaries['Salary_INR'].quantile(0.75)
salary_iqr = q3 - q1

print(f"Variance of Salaries: {salary_variance}")
print(f"Standard Deviation of Salaries: {salary_std_deviation}")
print(f"IQR of Salaries: {salary_iqr}")
```

```python
it_dpt_exp = df[df['Department'] == 'IT']
experience_range = it_dpt_exp['time_spent_company'].max() - it_dpt_exp['time_spent_company'].min()

print(f"Range of Experience in the IT Department: {experience_range} years")
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(figsize=(10, 6))
sb.boxplot(data=df, x='Department', y='time_spent_company')
plt.title('Experience by Department')
plt.xlabel('Department')
plt.ylabel('Experience (Years)')
plt.xticks(rotation=45)
plt.show()
```
